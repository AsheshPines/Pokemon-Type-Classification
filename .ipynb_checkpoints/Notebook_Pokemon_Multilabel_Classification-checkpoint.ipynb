# Libraries

```python
import numpy as np #For Loading Data, and array operations
import matplotlib.pyplot as plt # For Plotting Figures
```

# Data Loading from Kaggle Dataset

```python
#Load data from pokemon.csv, format: Name,Type1,Type2
#fills missing values
data = np.loadtxt('pokemonedited.csv', delimiter=',',skiprows=1, unpack=True, dtype = str)
X, y = data[0].T, data[1:] 
#X contains pokemon names
#y contains pokemon typing (class)
```

# Plot Type Charts

```python
#y[0] contains primary typing (type1), y[0] contains the secondary typing

#typing and frequency array
primary = ['Grass','Fire','Water','Bug','Normal','Poison','Electric','Ground','Fairy',
         'Fighting','Psychic','Rock','Ghost','Ice','Dragon','Dark','Steel','Flying']
type1 = np.zeros([18]).astype(np.int64)

#get the frequencies
for i in y[0]:
    if(i=='Grass'):
        type1[0] = type1[0] + 1
    if(i=='Fire'):
        type1[1] = type1[1] + 1
    if(i=='Water'):
        type1[2] = type1[2] + 1
    if(i=='Bug'):
        type1[3] = type1[3] + 1
    if(i=='Normal'):
        type1[4] = type1[4] + 1
    if(i=='Poison'):
        type1[5] = type1[5] + 1
    if(i=='Electric'):
        type1[6] = type1[6] + 1
    if(i=='Ground'):
        type1[7] = type1[7] + 1
    if(i=='Fairy'):
        type1[8] = type1[8] + 1
    if(i=='Fighting'):
        type1[9] = type1[9] + 1
    if(i=='Psychic'):
        type1[10] = type1[10] + 1
    if(i=='Rock'):
        type1[11] = type1[11] + 1
    if(i=='Ghost'):
        type1[12] = type1[12] + 1
    if(i=='Ice'):
        type1[13] = type1[13] + 1
    if(i=='Dragon'):
        type1[14] = type1[14] + 1
    if(i=='Dark'):
        type1[15] = type1[15] + 1
    if(i=='Steel'):
        type1[16] = type1[16] + 1
    if(i=='Flying'):
        type1[17] = type1[17] + 1


plt.figure(figsize=(20,12))
plt.bar(primary, type1)
plt.title("Type1 Frequency", fontsize=14)
plt.xlabel("Type1")
plt.ylabel("Frequency")
plt.show()
```

```python
#typing and frequency array
secondary = ['Grass','Fire','Water','Bug','Normal','Poison','Electric','Ground','Fairy',
         'Fighting','Psychic','Rock','Ghost','Ice','Dragon','Dark','Steel','Flying', 'notype']
type2 = np.zeros([19]).astype(np.int64)

#get the frequencies
for i in y[1]:
    if(i=='Grass'):
        type2[0] = type2[0] + 1
    if(i=='Fire'):
        type2[1] = type2[1] + 1
    if(i=='Water'):
        type2[2] = type2[2] + 1
    if(i=='Bug'):
        type2[3] = type2[3] + 1
    if(i=='Normal'):
        type2[4] = type2[4] + 1
    if(i=='Poison'):
        type2[5] = type2[5] + 1
    if(i=='Electric'):
        type2[6] = type2[6] + 1
    if(i=='Ground'):
        type2[7] = type2[7] + 1
    if(i=='Fairy'):
        type2[8] = type2[8] + 1
    if(i=='Fighting'):
        type2[9] = type2[9] + 1
    if(i=='Psychic'):
        type2[10] = type2[10] + 1
    if(i=='Rock'):
        type2[11] = type2[11] + 1
    if(i=='Ghost'):
        type2[12] = type2[12] + 1
    if(i=='Ice'):
        type2[13] = type2[13] + 1
    if(i=='Dragon'):
        type2[14] = type2[14] + 1
    if(i=='Dark'):
        type2[15] = type2[15] + 1
    if(i=='Steel'):
        type2[16] = type2[16] + 1
    if(i=='Flying'):
        type2[17] = type2[17] + 1
    if(i=='notype'):
        type2[18] = type2[18] + 1

plt.figure(figsize=(20,12))
plt.bar(secondary, type2)
plt.title("Type2 Frequency", fontsize=14)
plt.xlabel("Type2")
plt.ylabel("Frequency")
plt.show()

#display without notype 
#shows that flying is the predominant secondary type, but most Pokemon don't have a secondary type
plt.figure(figsize=(20,12))
plt.bar(secondary[0:18], type2[0:18])
plt.title("Type2 Frequency", fontsize=14)
plt.xlabel("Type2")
plt.ylabel("Frequency")
plt.show()
```

```python
#typing and frequency array
types = ['Grass','Fire','Water','Bug','Normal','Poison','Electric','Ground','Fairy',
         'Fighting','Psychic','Rock','Ghost','Ice','Dragon','Dark','Steel','Flying', 'notype']
#Sum of type1 and type2
type = np.zeros([19]).astype(np.int64)
type = np.append(type1,0) + type2

plt.figure(figsize=(20,12))
plt.bar(types, type)
plt.title("Type Frequency", fontsize=14)
plt.xlabel("Types")
plt.ylabel("Frequency")
plt.show()
```
